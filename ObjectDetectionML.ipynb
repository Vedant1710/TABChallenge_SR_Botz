```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
import glob
import PIL
from tensorflow.keras.preprocessing import image
```

```python
train_set=ImageDataGenerator(rescale=1/255)
test_set=ImageDataGenerator(rescale=1/255)

train_dataset = train_set.flow_from_directory("C:/Users/Vedant/Desktop/Auto Bot Challenge/Dataset/Train/",
                                          target_size=(112,112),
                                          batch_size = 32,
                                          class_mode = 'sparse')
                                         
test_dataset = test_set.flow_from_directory("C:/Users/Vedant/Desktop/Auto Bot Challenge/Dataset/Test/",
                                          target_size=(112,112),
                                          batch_size =32,
                                          class_mode = 'sparse')
```

```python
train_dataset.class_indices
```

```python
#Making cnn model
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(112, 112, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(1000, activation='relu'),
    layers.Dense(500, activation='relu'),
    layers.Dense(7, activation='sigmoid')
])
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(train_dataset, steps_per_epoch = 100, epochs = 10, validation_data = test_dataset)
```

```python
#Prediction n further usage
def predictImage(filename):
    img1 = image.load_img(filename,target_size=(112,112))
    plt.imshow(img1)
    Y = image.img_to_array(img1)
    X = np.expand_dims(Y,axis=0)
    val = cnn.predict(X)
    print(val)
    if val == 0:
        plt.xlabel("Animal",fontsize=30)
    elif val == 1:  
        plt.xlabel("Blue Light",fontsize=30)
    elif val == 2:  
        plt.xlabel("Green Light",fontsize=30)
    elif val == 3:  
        plt.xlabel("Person",fontsize=30)
    elif val == 4:  
        plt.xlabel("Red Light",fontsize=30)
    elif val == 5:  
        plt.xlabel("Road Cones",fontsize=30)
    elif val == 6:  
        plt.xlabel("Stop",fontsize=30)
```
